```python
import logging
import smart_open
import os
import gensim
import numpy as np
from numpy import loadtxt

def read_corpus(fname, tokens_only=False):
    with smart_open.open(fname, encoding="iso-8859-1") as f:
        for i, line in enumerate(f):
            tokens = gensim.utils.simple_preprocess(line)
            if tokens_only:
                yield tokens
            else:
                # For training data, add tags
                yield gensim.models.doc2vec.TaggedDocument(tokens, [i])
                
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
                
path = '../bped_data'
train_fr_dir = os.path.join(path, 'entrain')
test_fr_dir = os.path.join(path, 'entest')
# fr_train_file = os.path.join(train_fr_dir, 'IWSLT17.en-fr.tok.tc.bpe.fr')
# fr_test_file = os.path.join(test_fr_dir, 'IWSLT17.TED.dev2010.en-fr.tok.tc.bpe.fr')

fr_train_files = []
fr_test_files = []
```

```python
train_corpus = []
for file in os.listdir(train_fr_dir):
    fr_train_files.append(os.path.join(train_fr_dir, file))
    train_corpus += list(read_corpus(os.path.join(train_fr_dir, file)))

test_corpus = []
total = 0 
for file in os.listdir(test_fr_dir):
    print(filelen)
    fr_test_files.append(os.path.join(test_fr_dir, file))
    test_corpus += list(read_corpus(os.path.join(test_fr_dir, file), tokens_only=True))
    filelen = len(test_corpus)
    total+=filelen
    print(filelen)
print("total %d" % total)
```

```python
model = gensim.models.doc2vec.Doc2Vec(vector_size=50, min_count=2, epochs=40)

model.build_vocab(train_corpus)

model.train(train_corpus, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
import numpy as np
from numpy import loadtxt

test_en_data = []
for doc_id in range(len(test_corpus)):
    inferred_vector = model.infer_vector(test_corpus[doc_id])
    test_en_data.append(inferred_vector)
    
```
