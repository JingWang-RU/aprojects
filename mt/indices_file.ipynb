```python
import smart_open
import os

path = '/data'
index_dir = os.path.join(path, 'filtered')
method = 'Filter'
# method = 'RF'
# dev en
save_file = method + '_IWSLT17.TED.dev2010_en_data';
test_en_dir = os.path.join(path, 'entest')
deven = 'IWSLT17.TED.dev2010.en-fr.tok.tc.bpe.en';
deven1 = 'IWSLT17.TED.dev2010.en-fr.tok.tc.bped_1.en';
deven2 = 'IWSLT17.TED.dev2010.en-fr.tok.tc.bped_2.en';
deven3 = 'IWSLT17.TED.dev2010.en-fr.tok.tc.bped_3.en';

deven_ind = method + '_IWSLT17.TED.dev2010_en_index.txt'
indices = open(os.path.join(index_dir,deven_ind), "r").read().splitlines()

file0 = os.path.join(test_en_dir, deven)
lines0 = smart_open.open(file0, encoding="iso-8859-1").readlines()
file1 = os.path.join(test_en_dir, deven1)
lines1 = smart_open.open(file1, encoding="iso-8859-1").readlines()
file2 = os.path.join(test_en_dir, deven2)
lines2 = smart_open.open(file2, encoding="iso-8859-1").readlines()
file3 = os.path.join(test_en_dir, deven3)
lines3 = smart_open.open(file3, encoding="iso-8859-1").readlines()

if os.path.exists(os.path.join(index_dir,save_file)):
    os.remove(os.path.join(index_dir,save_file))
else:
    print("The file does not exist")
savefid = open(os.path.join(index_dir,save_file), "w+",encoding="iso-8859-1")
j = 0
for i in range(len(indices)):
    if (float(indices[i].strip())) == 1:
        savefid.write(lines0[j])
    elif (float(indices[i].strip())) == 2:
        savefid.write(lines1[j])
    elif (float(indices[i].strip())) == 3:
        savefid.write(lines2[j])
    else:
        savefid.write(lines3[j])
    j = j+1
savefid.close()
```
